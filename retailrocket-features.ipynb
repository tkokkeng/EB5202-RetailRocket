### Pagetime

```python
pages_more_than1 = sessions_df['pages'] > 1    
```

```python
sessions_df.loc[pages_more_than1, 'pagetime'] = (sessions_df.loc[pages_more_than1, 'enddate'] - sessions_df.loc[pages_more_than1, 'startdate']) /\
                                                (sessions_df.loc[pages_more_than1, 'pages'] - 1)
```

```python
pages_less_than1 = pages_more_than1.apply(lambda x: not x)
```

```python
sessions_df.loc[pages_less_than1, 'pagetime'] = pd.Timedelta(0)
```

### Diffcat

```python
infile = open(os.path.join('data', 'itemtocat.pkl'), 'rb')
itemtocat_dict = pickle.load(infile)
infile.close()
```

```python
itemtocat_dict[177677]
```

There are itemids which are not found in the dictionary. Need to handle this when converting itemid to categoryid.

```python
sessions_df.loc[10, :]
```

```python
for i in sessions_df['view'][10]:
    print(itemtocat_dict[i])
```

```python
def to_categoryid(itemid, enddate):
    catid = ''
    
    for time, cat in itemtocat_dict.get(itemid, []):
        if enddate <= time:
            catid = cat
            break
    if not catid:
        if itemtocat_dict.get(itemid, []):  # enddate did not match
            catid = itemtocat_dict[itemid][-1][1]  # set categiryid to last value
        else:
            catid = str(itemid) + '_no_cat_id'  # itemid not in dictionary
    return catid
```

```python
{ to_categoryid(i, sessions_df['enddate'][10]) for i in sessions_df['view'][10] }
```

Given the session enddate of 2015-05-14, the categoryid 642 is correct.

```python
sessions_df['diffcat'] = sessions_df.loc[:, ['enddate', 'view']].apply(
    lambda x: len({ to_categoryid(i, x['enddate']) for i in x['view'] }) / len(x['view']) if x['view'] else 0, axis=1)
#sessions_df['diffcat'] = sessions_df.loc[:, ['enddate', 'view']].apply(lambda x: (x.view), axis=1)
#sessions_df.loc[[0, 1, 2], ['enddate', 'view']].apply(lambda x: print(x['view']), axis=1)
```

### Diffprod

```python
sessions_df['diffprod'] = sessions_df['view'].map(lambda x: len(set(x)) / len(x) if x else 0)
```

```python
sessions_df.head()
```

diffprod is 4 / 8 = .5

```python
sessions_df['diffprod'].to_frame().info(null_counts=True)
```

```python
sessions_df['diffprod'].describe()
```

### Prodcat

```python
sessions_df['prodcat'] = sessions_df.loc[:, ['enddate', 'view']].apply(
    lambda x: len(set(x['view'])) / len({ to_categoryid(i, x['enddate']) for i in x['view'] }) if x['view'] else 0, axis=1)
```

```python
sessions_df.head()
```

For visitorid 2, previously unique categoryid was determined to be 2 (see diffcat), the unique itemid was 4. prodcat is 4 / 2 = 2 which is correct in the above table.

```python
sessions_df['prodcat'].to_frame().info(null_counts=True)
```

```python
sessions_df['prodcat'].describe()
```

### Maxrep

```python
sessions_df['maxrep'] = sessions_df['view'].map(lambda x: Counter(x).most_common(1)[0][1] if x else 0)
```
