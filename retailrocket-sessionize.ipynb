## Transform data

```python
# convert unix time to pandas datetime
events_df['date'] = pd.to_datetime(events_df['timestamp'], unit='ms', origin='unix')
```

```python
events_df.head()
```

```python
# convert event to categorical
events_df['event_type'] = events_df['event'].astype('category')
```
